```python
def add_target(team):
    team['target'] = team['won'].shift(-1)
    return team
```

```python
data = data.groupby('team', group_keys=False).apply(add_target)
```

```python
# feature selection
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rc = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rc, n_features_to_select=30, direction='forward', cv=split)
```

```python
removed_cols = ['season', 'date', 'won', 'target', 'team', 'team_opp']
```

```python
selected_cols = data.columns[~data.columns.isin(removed_cols)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[selected_cols] = scaler.fit_transform(data[selected_cols])
```

```python
data.head()
```

```python
sfs.fit(data[selected_cols], data['target'])
```

```python
predictors = list(selected_cols[sfs.get_support()])
```

```python
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data['season'].unique())
    for i in range(start, len(seasons), step):
        season = seasons[i]

        train = data[data['season'] < season]
        test = data[data['season'] == season]

        model.fit(train[predictors], train['target'])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        
        combined = pd.concat([test['target'], preds], axis=1)
        combined.columns = ['actual', 'prediction']

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(data, rc, predictors)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score

predictions = predictions[predictions['actual'] != 2]
accuracy_score(predictions['actual'], predictions['prediction'])
```

```python
data.groupby('home').apply(lambda x: x[x['won'] == 1].shape[0] / x.shape[0])
```

```python
data_rolling = data[list(selected_cols) + ['won','team','season']]
```

```python
def find_team_averages(team):
    team = team.select_dtypes(include=[float,int])
    rolling = team.rolling(10).mean()
    return rolling

data_rolling = data_rolling.groupby(['team', 'season'], group_keys=False).apply(find_team_averages)
```

```python
data_rolling
```

```python
rolling_cols = [f'{col}_10' for col in data_rolling.columns]
data_rolling.columns = rolling_cols

data = pd.concat([data, data_rolling], axis=1)
```

```python
data
```

```python
data = data.dropna()
```

```python
def shift_col(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col
 
def add_col(data, col_name):
    return data.groupby('team', group_keys=False).apply(lambda x: shift_col(x,col_name))

data['home_next'] = add_col(data, 'home')
data['team_opp_next'] = add_col(data, 'team_opp')
data['date_next'] = add_col(data, 'date')
```

```python
data
```

```python
data = data.copy()
```

```python
full = data.merge(data[rolling_cols + ['team_opp_next', 'date_next', 'team']], 
                  left_on=['team', 'date_next'], 
                  right_on=['team_opp_next', 'date_next'])
```
